剑指 Offer 24. 反转链表
定义一个函数，输入一个链表的头节点，反转该链表并输出反转后链表的头节点。

示例:
输入: 1->2->3->4->5->NULL
输出: 5->4->3->2->1->NULL

```python
class ListNode(object):
    def __init__(self, x):
        self.val = x
        self.next = None
def createListNode(arr):
    head=ListNode(arr[0])
    current=head
    for i in range(1,len(arr)):
        current.next=ListNode(arr[i])
        current=current.next
    return head
given_list=[1,2,3,4,5]
head=createListNode(given_list)
```

```python
#递归
#使用递归函数，一直递归到链表的最后一个结点，该结点就是反转后的头结点，记作 ret
#此后，每次函数在返回的过程中，让当前结点的下一个结点的 next 指针指向当前节点。
#同时让当前结点的 next指针指向 NULL，从而实现从链表尾部开始的局部反转
#当递归函数全部出栈后，链表反转完成。

class Solution(object):
    def reverseList(self, head):
        if head is None or head.next is None:
            return head
        rets = self.reverseList(head.next)#遍历到了链表的终点
        head.next.next = head
        head.next = None
        return rets#rets屁股后面的每一个元素都做了反转，它本身并不需要知道
```

```python
class Solution(object):
    def reverseList(self, head):
        if head is None:
            return None
        cur = head
        while head.next is not None:
            t = head.next.next
            head.next.next = cur
            cur = head.next
            head.next = t
        return cur
#执行用时：40 ms, 在所有 Python3 提交中击败了91.80%的用户
#内存消耗：17.9 MB, 在所有 Python3 提交中击败了12.03%的用户
```
